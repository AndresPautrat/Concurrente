```
%%writefile counter.go
package main
import (
    "fmt"
    "math/rand"
    "time"
)

var n int

func P(){
    var temp int
    for i := 0; i<10; i++{
        rndPause()
        temp=n
        rndPause()
        n=temp+1
    }
}

func main(){
    rand.Seed(time.Now().UTC().UnixNano())
    go P()
    go P()
    time.Sleep(2*time.Second)
    fmt.Println(n)
}

func rndPause(){
    dur := time.Duration(rand.Intn(50)+100)
    time.Sleep(dur*time.Millisecond)
}
```
